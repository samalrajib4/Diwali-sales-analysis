**Analysis base on 'Gender & Age_group'**

```python
plt.figure(figsize=(3,4))
count_plt = sns.countplot(data=df, x='Gender')
plt.bar_label(count_plt.containers[0], label_type='edge')
plt.title("Purchase based on 'Gender'")
plt.show()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(3,4))
sales_gender = sns.barplot(data=sales_gen, x='Gender', y='Amount')
plt.bar_label(sales_gender.containers[0], label_type='edge', fmt='%.2f')
plt.title("Sales Based on 'Gender'")
plt.show()
```

From the above graph we can see that most of the byers are females and the purchasing power of female are also greater than men.

```python
sns.countplot(data=df, x='Age Group', hue='Gender')
plt.title("Sales based on 'Age_group' of Male & Female")
plt.ylabel('Count')
plt.show()
```

```python
sales_age_grp = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(8,5))
age_grp = sns.barplot(data=sales_age_grp, x='Age Group', y='Amount')
plt.bar_label(age_grp.containers[0], label_type='edge', fmt='%.2f',)
plt.title("Sales based on 'Age_group' of Male & Female")
plt.show()
```

From above graphs we can see that most of the byer are of age group between 26-35 yrs

**Analysis base on 'States'**

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(15,5))
bar_plot = sns.barplot(data=sales_state,x='State',y='Orders')
plt.bar_label(bar_plot.containers[0], label_type='edge')
plt.title("Total Orders based on 'States'")

plt.xticks(rotation=15)
plt.show()
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize=(15,5))
bar_plot = sns.barplot(data=sales_state,x='State',y='Amount')
plt.bar_label(bar_plot.containers[0], fmt='%.2f', label_type='edge')
plt.title("Total Saales based on 'States'")

plt.xticks(rotation=15)
plt.show()
```

From the above graphs we can see that most of the orders & total sales/amount are Uttar Pradesh, Maharashtra and Karnataka respectively.

**Analysis base on 'Marital_Status'**

```python
plt.figure(figsize=(3,4))
bar_plt=sns.countplot(data=df, x='Marital_Status')
plt.bar_label(bar_plt.containers[0], label_type='edge')
plt.title("Sales based on 'Marital_Status' of the customers")
```

```python
plt.figure(figsize=(6,4))
bar_plt=sns.countplot(data=df, x='Marital_Status', hue='Gender')
plt.bar_label(bar_plt.containers[0], label_type='edge')
plt.bar_label(bar_plt.containers[1], label_type='edge')
plt.title("Sales based on 'Marital_Status' of the customers")
```

From the above graph we can see the most of the byers are married(Women) and they have high purchasing power.

**Analysis base on 'Occupation'**

```python
plt.figure(figsize=(14,5))

occu_p = sns.countplot(data=df,x='Occupation')
plt.title("Orders base on 'Occupation'")
plt.ylabel('Order')
plt.bar_label(occu_p.containers[0])
plt.xticks(rotation=18)

plt.show()
```

```python
occu_sls = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
plt.figure(figsize=(14,9))

occu_plt = sns.barplot(data=occu_sls, x='Occupation', y='Amount')
plt.title("Sales base on 'Occupation'")
plt.bar_label(occu_plt.containers[0], rotation=70, label_type='center', fmt='%.2f')
plt.xticks(rotation=18)
plt.ylabel('Sales')
plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector.

**Analysis base on 'Product_Category'**

```python
plt.figure(figsize=(18,5))

pdt_c = sns.countplot(data=df,x='Product_Category')
plt.title("Orders base on 'Product_Category'")
plt.ylabel('Order')
plt.bar_label(pdt_c.containers[0])
plt.xticks(rotation=22)

plt.show()
```

```python
pdt_cat = df.groupby (['Product_Category'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
plt.figure(figsize=(20,6))
cat_plt = sns.barplot(data=pdt_cat, x='Product_Category', y='Amount')
plt.bar_label(cat_plt.containers[0],fmt='%.2f', rotation= 4)
plt.xticks(rotation=20)
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category.

**Conclusion**

From the above analysis found, married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.
